```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import relax as re
```

```python
f1 = lambda x: np.pi**2*np.sin(np.pi*x)
n1 = 100
```

```python
list(re.methods.keys())
```

```python
fig, ax = plt.subplots(3,1)
for m in re.methods.keys():
    u, num, r = re.Poisson1D(f1, n1, method=m)
    ax[0].plot(u)
    ax[1].plot(r)
    ax[2].plot(r)
    ax[2].set_yscale('log')
ax[0].set_title('solution')
ax[1].set_title('residue')
ax[2].set_title('log residue')
ax[2].legend(list(re.methods.keys()))
for i in range(3):
    s = np.array(ax[i].get_position()).flatten()
    print(s)
    ax[i].set_position([s[0], s[1], s[2]-s[0], s[3]-s[1]-0.05])
```
